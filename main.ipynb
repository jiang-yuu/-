```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import cifar10
import os

# 1. 
def load_and_preprocess_data():
    (x_train, y_train), (x_test, y_test) = cifar10.load_data()

    # float16
    x_train, x_test = x_train.astype(np.float16) / 255.0, x_test.astype(np.float16) / 255.0

    #  (32 * 32 * 3)
    x_train, x_test = x_train.reshape(x_train.shape[0], -1), x_test.reshape(x_test.shape[0], -1)

    # One-hot 
    y_train, y_test = np.eye(10)[y_train.reshape(-1)], np.eye(10)[y_test.reshape(-1)]

    return x_train, y_train, x_test, y_test

# 2. 

# ReLU 
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)

# Sigmoid 
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sigmoid_x = sigmoid(x)
    return sigmoid_x * (1 - sigmoid_x)

# Tanh 
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2



# 3. 
class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size, activation_fn, activation_derivative_fn):
        #  Xavier 
        self.W1 = np.random.randn(input_size, hidden_size) * np.sqrt(2. / input_size)
        self.b1 = np.zeros((1, hidden_size))
        self.W2 = np.random.randn(hidden_size, output_size) * np.sqrt(2. / hidden_size)
        self.b2 = np.zeros((1, output_size))
        
        # 
        self.activation_fn = activation_fn
        self.activation_derivative_fn = activation_derivative_fn
    
    def forward(self, X):
        self.Z1 = np.dot(X, self.W1) + self.b1
        self.A1 = self.activation_fn(self.Z1)
        self.Z2 = np.dot(self.A1, self.W2) + self.b2
        self.A2 = self.softmax(self.Z2)
        return self.A2

    def backward(self, X, Y, learning_rate, lambda_reg):
        m = X.shape[0]
        
        # 
        dZ2 = self.A2 - Y
        dW2 = np.dot(self.A1.T, dZ2) / m + (lambda_reg / m) * self.W2
        db2 = np.sum(dZ2, axis=0, keepdims=True) / m
        
        # 
        dA1 = np.dot(dZ2, self.W2.T)
        dZ1 = dA1 * self.activation_derivative_fn(self.Z1)
        dW1 = np.dot(X.T, dZ1) / m + (lambda_reg / m) * self.W1
        db1 = np.sum(dZ1, axis=0, keepdims=True) / m
        
        # 
        self.W1 -= learning_rate * dW1
        self.b1 -= learning_rate * db1
        self.W2 -= learning_rate * dW2
        self.b2 -= learning_rate * db2
    
    def softmax(self, Z):
        eZ = np.exp(Z - np.max(Z, axis=1, keepdims=True))  # 
        return eZ / np.sum(eZ, axis=1, keepdims=True)
    
    def compute_loss(self, Y_pred, Y_true, lambda_reg):
        m = Y_true.shape[0]
        #  + L2 
        cross_entropy_loss = -np.sum(Y_true * np.log(Y_pred + 1e-8)) / m
        l2_loss = (lambda_reg / (2 * m)) * (np.sum(np.square(self.W1)) + np.sum(np.square(self.W2)))
        return cross_entropy_loss + l2_loss

# 4. 
def train_model(model, X_train, Y_train, X_val, Y_val, num_epochs=10, batch_size=64, learning_rate=0.001, lambda_reg=0.01, learning_rate_decay=False):
    train_loss_history, val_loss_history, val_accuracy_history = [], [], []

    best_val_loss = float('inf')
    best_weights = None
    
    for epoch in range(num_epochs):
        num_batches = X_train.shape[0] // batch_size
        for i in range(num_batches):
            # 
            batch_start = i * batch_size
            batch_end = batch_start + batch_size
            X_batch, Y_batch = X_train[batch_start:batch_end], Y_train[batch_start:batch_end]
            
            # 
            Y_pred = model.forward(X_batch)
            
            # 
            model.backward(X_batch, Y_batch, learning_rate, lambda_reg)
        
        # 
        Y_train_pred, Y_val_pred = model.forward(X_train), model.forward(X_val)
        train_loss = model.compute_loss(Y_train_pred, Y_train, lambda_reg)
        val_loss = model.compute_loss(Y_val_pred, Y_val, lambda_reg)

        # 
        accuracy = np.mean(np.argmax(Y_val_pred, axis=1) == np.argmax(Y_val, axis=1))
        
        train_loss_history.append(train_loss)
        val_loss_history.append(val_loss)
        val_accuracy_history.append(accuracy)
        
        # 
        if learning_rate_decay:
            learning_rate *= 0.9
        
        # 
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_weights = (model.W1.copy(), model.b1.copy(), model.W2.copy(), model.b2.copy())
        
        print(f'Epoch {epoch+1}/{num_epochs} - Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Val Accuracy: {accuracy:.4f}')
    
    # 
    model.W1, model.b1, model.W2, model.b2 = best_weights
    
    return train_loss_history, val_loss_history, val_accuracy_history

# 5. 
def test_model(model, X_test, Y_test):
    Y_test_pred = model.forward(X_test)
    accuracy = np.mean(np.argmax(Y_test_pred, axis=1) == np.argmax(Y_test, axis=1))
    print(f'Test Accuracy: {accuracy:.4f}')
    return accuracy

# 6. 
def save_model_weights(model, filename):
    np.savez(filename, W1=model.W1, b1=model.b1, W2=model.W2, b2=model.b2)
    print(f'Model weights saved to {filename}')

# 7. 
def plot_training_progress(train_loss_history, val_loss_history, val_accuracy_history):
    """
    
    
    :param train_loss_history: 
    :param val_loss_history: 
    :param val_accuracy_history: 
    """
    plt.figure(figsize=(12, 6))

    # 
    plt.subplot(1, 2, 1)
    plt.plot(train_loss_history, label='Train Loss')
    plt.plot(val_loss_history, label='Validation Loss')
    plt.title('Loss Curve')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # 
    plt.subplot(1, 2, 2)
    plt.plot(val_accuracy_history, label='Validation Accuracy')
    plt.title('Accuracy Curve')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

# 8. 
def hyperparameter_search(X_train, Y_train, X_val, Y_val, X_test, Y_test):
    learning_rates = [0.001, 0.01, 0.1]
    hidden_sizes = [64, 128, 256]
    lambda_regs = [0.001, 0.01, 0.1]
    
    best_accuracy = 0
    best_params = None
    best_val_loss = float('inf')
    best_train_loss = float('inf')
    best_model = None  # 
    
    # 
    all_accuracies = []
    all_params = []
    
    for lr in learning_rates:
        for hidden_size in hidden_sizes:
            for lambda_reg in lambda_regs:
                print(f'Training with lr={lr}, hidden_size={hidden_size}, lambda_reg={lambda_reg}')
                
                model = NeuralNetwork(input_size=3072, hidden_size=hidden_size, output_size=10, activation_fn=relu, activation_derivative_fn=relu_derivative)
                
                # 
                train_loss_history, val_loss_history, val_accuracy_history = train_model(
                    model, X_train, Y_train, X_val, Y_val, num_epochs=30, learning_rate=lr, lambda_reg=lambda_reg)
                
                # 
                accuracy = test_model(model, X_test, Y_test)
                
                if accuracy > best_accuracy:
                    best_accuracy = accuracy
                    best_params = (lr, hidden_size, lambda_reg)
                    best_train_loss = train_loss_history[-1]
                    best_val_loss = val_loss_history[-1]
                    best_model = model  
                
                # 
                all_params.append((lr, hidden_size, lambda_reg))
                all_accuracies.append(accuracy)
    
    # 
    print(f'\nBest Parameters:')
    print(f'Learning Rate = {best_params[0]}')
    print(f'Hidden Size = {best_params[1]}')
    print(f'Lambda Regularization = {best_params[2]}')
    print(f'Best Test Accuracy = {best_accuracy:.4f}')
    print(f'Best Validation Loss = {best_val_loss:.4f}')
    print(f'Best Train Loss = {best_train_loss:.4f}')
    
    # 
    plot_hyperparameter_accuracy(all_params, all_accuracies)
    
    # 
    plot_training_progress(best_model.train_loss_history, best_model.val_loss_history, best_model.val_accuracy_history)
    
    # 
    plot_best_model_parameters(best_model)
    
    # 
    save_model_weights(best_model, 'best_model_weights.npz')

def plot_hyperparameter_accuracy(params, accuracies):
    # 
    params = np.array(params)
    accuracies = np.array(accuracies)

    # DataFrame
    import pandas as pd
    df = pd.DataFrame(params, columns=['Learning Rate', 'Hidden Size', 'Lambda Reg'])
    df['Accuracy'] = accuracies

    # 
    pivot_df = df.pivot_table(index='Learning Rate', columns='Hidden Size', values='Accuracy', aggfunc='max')
    
    # 
    plt.figure(figsize=(10, 8))
    sns.heatmap(pivot_df, annot=True, cmap='viridis', fmt=".4f", linewidths=0.5)
    plt.title('Hyperparameter Tuning: Accuracy vs Learning Rate & Hidden Size')
    plt.xlabel('Hidden Size')
    plt.ylabel('Learning Rate')
    plt.show()


# 10. 
def plot_best_model_parameters(model):
    # 
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.heatmap(model.W1, cmap="viridis", cbar=True)
    plt.title("First Layer Weights (W1)")
    plt.xlabel("Hidden Neurons")
    plt.ylabel("Input Features")

    # 
    plt.subplot(1, 2, 2)
    sns.heatmap(model.W2, cmap="viridis", cbar=True)
    plt.title("Second Layer Weights (W2)")
    plt.xlabel("Output Neurons")
    plt.ylabel("Hidden Neurons")
    
    plt.tight_layout()
    plt.show()

    # 
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.hist(model.b1.flatten(), bins=50, color='blue', alpha=0.7)
    plt.title("Bias of First Layer (b1)")
    plt.xlabel("Bias Values")
    plt.ylabel("Frequency")

    plt.subplot(1, 2, 2)
    plt.hist(model.b2.flatten(), bins=50, color='green', alpha=0.7)
    plt.title("Bias of Second Layer (b2)")
    plt.xlabel("Bias Values")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

# 9. 
if __name__ == '__main__':
    # 
    x_train, y_train, x_test, y_test = load_and_preprocess_data()
    
    # 
    hyperparameter_search(x_train, y_train, x_test, y_test, x_test, y_test)
```

```python
def visualize_hidden_neuron_features(model, neuron_indices=[0, 5, 10], input_shape=(32, 32, 3)):
    """
    

    :param model: 
    :param neuron_indices: 
    :param input_shape:  (32, 32, 3)
    """
    W1 = model.W1  # shape: (3072, hidden_size)

    plt.figure(figsize=(len(neuron_indices) * 3, 3))

    for i, neuron_index in enumerate(neuron_indices):
        weights = W1[:, neuron_index]  # shape: (3072,)
        image = weights.reshape(input_shape)  # reshape to (32, 32, 3)
        
        #  0~1 
        image = (image - image.min()) / (image.max() - image.min() + 1e-8)

        plt.subplot(1, len(neuron_indices), i + 1)
        plt.imshow(image)
        plt.axis('off')
        plt.title(f'Neuron {neuron_index}')

    plt.suptitle("Visualization of Hidden Layer Neuron Features")
    plt.tight_layout()
    plt.show()
visualize_hidden_neuron_features(best_model, neuron_indices=[0, 3, 10, 25, 50])
```
